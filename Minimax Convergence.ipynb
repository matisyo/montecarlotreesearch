```python
from game.game import *
from memorymontecarlo.memcts import Memcts
from minmax.minimax import *
from montecarlo.uct import uct
```

```python
LARGO = 3
TABLERO_ALTO = 4
TABLERO_ANCHO = 3
t = Connect4(TABLERO_ANCHO, TABLERO_ALTO, LARGO)
```

```python
agente = MiniMax(t,12)
agente.best_move()
```

```python
agentemmc = Memcts(t)
r = []
for i in range(3,59):
    a = agentemmc.uct(i, plot=False,persist=True)
    tmp = list(map(lambda x:float(x[1:-1].split(',')[0]) ,a[1].split(" ")[1::2]))
    r.append(tmp)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
ax = pd.DataFrame(r).plot(figsize=(8,7))
ax.set_title( "MCTS to Minimax Convergence")
ax.set_xlabel("MCTS Iters")
ax.set_ylabel("Reward")
ax.legend(title="Actions")
ax.set_xticklabels([i*(i+1)//2 for i in ax.get_xticks()])
None
```
